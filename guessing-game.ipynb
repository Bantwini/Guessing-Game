## <span style="background-color: #ADD8E6; padding: 5px; border-radius: 5px;">ğŸ“œ Rules:</span>

1. ğŸ² **The computer randomly selects a number between 0 and 100.**
2. ğŸ–Šï¸ **The player is asked to input the number of attempts they want (between 0 and 10).**
3. ğŸ” **The player then guesses the number within the specified attempts.**
4. ğŸ“‰ **After each guess, the player is informed if the guess is too high or too low.**
5. ğŸ”„ **The game continues until the player guesses the correct number or runs out of attempts.**
6. ğŸ† **Scoring:**
   - The player starts with a score of 100 points.
   - For each incorrect guess, 10 points are deducted.
   - The final score is displayed when the player guesses the correct number or runs out of attempts.


```python
import random

number = random.randint(0, 100)

while True:
    try:
        attempts = int(input("How many attempts would you like (enter a number between 0 and 5): "))
        if 0 <= attempts <= 5:
            break
        else:
            print("Invalid input. Please enter a number between 0 and 5")
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

remaining_attempts = attempts
attempt_count = 0
score = 100  # Starting score

while remaining_attempts > 0:
    try:
        guess = int(input("Enter your guess: "))
    except ValueError:
        print("Invalid input. Please enter a numerical value.")
        continue

    remaining_attempts -= 1
    attempt_count += 1

    if guess == number:
        print(f"Congratulations! You guessed it in {attempt_count} attempts. Your score: {score}")
        break
    elif guess > number:
        print("Lower")
    else:
        print("Higher")

    score -= 10  # Decrease score by 10 for each incorrect guess

    if remaining_attempts == 0:
        print(f"Your attempts are over. The correct number was: {number}")
```
